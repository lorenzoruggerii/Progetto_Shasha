## Filtering the edges in df_edges

The problem is that we have to add edges with nodes representing the states of the agents (sick/healthy/recovered)

```python
import numpy as np
from random import sample
from random import choices
from random import randrange
import random
import pandas as pd

from os.path import join, exists
```

```python
# Dataframe loading
folder_path = r"../Shasha/tNodeEmbed/data/Meetings"
nodes_path = join(folder_path, "nodes_1_prova.csv")
edges_path = join(folder_path, "edges_1_prova.csv")

nodes = pd.read_csv(nodes_path, index_col = False)
edges = pd.read_csv(edges_path, index_col = False)
```

## Add edges with respect to the states

```python
for identifier in nodes["ID"].unique():
    agent_states = nodes[nodes["ID"]==identifier]
    last_state = agent_states.iloc[0]["State"]
    last_datetime = agent_states.iloc[0]["Datetime"]
    for (i, row) in agent_states.iterrows():
        if row["State"] != last_state:
            new_edge = {"ID_1": identifier, "ID_2": last_state, "TS_start": last_datetime, "TS_end": row["Datetime"], "Type": "STATUS", "Edge_ID": "STATUS"}
            edges.loc[len(edges)] = new_edge
            last_state = row["State"]
            last_datetime = row["Datetime"]
            
    
```
